```python
import sqlite3

conn = sqlite3.connect("../sql-database/options_data.db")

# Total rows in enriched_options
total = conn.execute("SELECT COUNT(*) FROM enriched_options;").fetchone()[0]

# Rows still missing a BS price
pending = conn.execute("""
    SELECT COUNT(*) 
    FROM enriched_options 
    WHERE black_scholes_model_price IS NULL
""").fetchone()[0]

conn.close()

print(f"Total rows: {total:,}")
print(f"Rows pending BS price: {pending:,}")
```

```python
import pandas as pd

conn = sqlite3.connect("../sql-database/options_data.db")

# assume df has columns: 'ask', 'black_scholes_model_price'
df = pd.read_sql_query("SELECT contract,underlying, ask, black_scholes_model_price, stock_price_today, stock_price_contract_close FROM enriched_options where quote_date < '2013-02-01 00:00:00' limit 100000", conn)

# 1) Raw and pct deviation
df["diff"]     = df["ask"] - df["black_scholes_model_price"]
df["pct_diff"] = df["diff"] / df["black_scholes_model_price"] * 100

# 2) Flags
TH = 5.0  # 5% threshold
df["good_bid"]    = df["pct_diff"] <= -TH
df["overpriced"]  = df["pct_diff"] >= TH

# 3) Inspect
print("Good bids (ask ≥5% below model):")
print(df[df["good_bid"]].head())

print("\nOverpriced (ask ≥5% above model):")
print(df[df["overpriced"]].head())
```
